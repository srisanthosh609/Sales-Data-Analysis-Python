```python
import pandas as pd
import numpy as np

data = pd.read_csv(r"D:\data analysis\python\sales_data_sample.csv", encoding='unicode_escape')
```

```python
print(data.isnull().sum())
data = data.fillna(0)
```

```python
data = data.drop_duplicates()
```

```python
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'])
```

```python
data.head()
```

```python
data['TOTALSALES'] = data['QUANTITYORDERED'] * data['PRICEEACH']
```

```python
print(data.columns.tolist)
```

```python
data.head()
```

```python
Totalrevenue = data['TOTALSALES'].sum().round(2)
print("Total Revenue:", Totalrevenue)
```

```python
average_order_value =data['TOTALSALES'].mean().round(2)
print("average_order_value",average_order_value)
```

```python
Top_5_orders = data.sort_values(by='TOTALSALES', ascending=False).head(5)
print("Top 5 Orders:")
print(Top_5_orders[['ORDERNUMBER', 'TOTALSALES']])
```

```python
best_seller = data.groupby('PRODUCTLINE')['QUANTITYORDERED'].sum().sort_values(ascending=False).head(5)
print(best_seller.head())
best_seller=best_seller.idxmax()
print("Best-selling product line:", best_seller)
```

```python
monthly_sales = data.set_index('ORDERDATE')['TOTALSALES'].resample('MS').sum().reset_index()
monthly_sales.head()
```

```python
sales_array = data['TOTALSALES'].values
print(sales_array)
```

```python
mean_sales = np.mean(sales_array)
std_sales = np.std(sales_array)
max_sales = np.max(sales_array)
min_sales = np.min(sales_array)
print("Mean:", round(mean_sales, 2))
print("Standard Deviation:", round(std_sales, 2))
print("Maximum:", max_sales)
print("Minimum:", min_sales)
```

```python
sales_filtered = np.where(sales_array < 500, 0, sales_array)
print(sales_filtered)
```

```python
z_scores = (sales_array - np.mean(sales_array)) / np.std(sales_array)
outliers = data[np.abs(z_scores) > 3]
```

```python
print(z_scores)
print("Outliers:")
print(outliers[['ORDERNUMBER', 'TOTALSALES']])
```
